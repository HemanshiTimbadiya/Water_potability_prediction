### <b><span style='color:#338aff'>Step 4.1.1 |</span><span style='color:#338aff'> Check the distribution of continuous variables</span></b>


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Create a figure for the continuous variable distribution
plt.figure(figsize=(15, 20))  # Adjust the figure size if needed
for i, col in enumerate(data.columns):
    if col != 'Potability':
        ax = plt.subplot(5, 2, i + 1)  # Adjust the grid size as necessary
        sns.histplot(data[col], kde=True, color='blue')
        plt.title(f'Distribution of {col}', fontsize=14)
        plt.xlabel(col)
        plt.ylabel('Density')
        
        # Set the background color for the grid
        ax.set_facecolor('#f0f0fa')  # Change to your desired color

# Adjust layout
plt.tight_layout()
# # Save the figure before displaying it
# plt.savefig('chart_name.png', dpi=300, bbox_inches='tight', pad_inches=0)

plt.show()
```

__`pH`__: The distribution is nearly normal, with values mostly centered around 7-8. The mean pH is approximately 7.08, with some outliers on both acidic and basic sides, indicating potential water quality concerns at extremes.

__`Hardness`__: Water hardness is moderately right-skewed, with most values between 175-225 mg/L. The mean hardness is around 196.37 mg/L, showing moderate water hardness, with some samples exceeding typical levels.

__`Total Dissolved Solids (TDS)`__: The distribution is heavily skewed to the right. Most values are concentrated around 15,000-30,000 mg/L, with the mean at approximately 22,014 mg/L. Some extreme values indicate poor water quality in certain areas.

__`Chloramines`__: Chloramines show a normal distribution, with most values concentrated between 6-8 mg/L. The mean is 7.12 mg/L, reflecting effective disinfection, though higher concentrations are observed in some cases.

__`Sulfate`__: The distribution is approximately normal, with most sulfate levels ranging between 300-400 mg/L. The mean is 333.78 mg/L, indicating that sulfate concentrations are generally within acceptable limits, though there is some variability.

__`Conductivity`__: Conductivity is right-skewed, with most values between 350-450 ŒºS/cm. The mean conductivity is 426.21 ŒºS/cm, with a few higher values suggesting higher levels of dissolved ions.

__`Organic Carbon`__: Organic carbon shows a relatively normal distribution, with values concentrated around 10-18 ppm. The mean is 14.28 ppm, indicating typical organic content levels in water, though some samples have elevated organic matter.

__`Trihalomethanes`__: This feature is slightly right-skewed, with most values between 50-80 ppm. The mean concentration is 66.40 ppm, with some higher outliers, potentially indicating concerns over disinfection byproducts.

### <b><span style='color:#338aff'>Step 4.2.1 |</span><span style='color:#338aff'>Check the distribution with target variable (using boxplot)</span></b>  

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your DataFrame
num_features = len(data.columns) - 1  # Exclude the target variable 'Potability'
fig, axes = plt.subplots(nrows=(num_features + 1) // 2, ncols=2, figsize=(15, 20))
axes = axes.flatten()
features = data.columns[:-1]  # Exclude the target variable 'Potability'

# Define a color palette
palette = sns.color_palette("Set2", n_colors=2)

# Create box plots for each feature
for i, feature in enumerate(features):
    sns.boxplot(x='Potability', y=feature, data=data, ax=axes[i], hue='Potability', palette=palette, dodge=False)
    axes[i].set_title(f'Box Plot of {feature} by Potability', fontsize=14)
    axes[i].set_xlabel('Potability')
    axes[i].set_ylabel(feature)

    # Set x-tick labels directly
    axes[i].set_xticks([0, 1])  # Ensure there are ticks at the correct positions
    axes[i].set_xticklabels(['Non-Potable (0)', 'Potable (1)'])

# Hide any unused subplots (if the number of features is odd)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
# Save the figure as an image file
# plt.savefig('box_plots_by_potability.png', dpi=300)  # Adjust the dpi as needed
plt.show()
```

  <h4>Conclusion on Addressing Class Imbalance:</h4>

  <ul>
    <li>The analysis of the dataset reveals a significant class imbalance, with fewer samples categorized as <strong>Potable</strong> (1) compared to <strong>Not Potable</strong> (0).</li>
    <li>This imbalance can lead to biased predictions from machine learning models, favoring the majority class, which may compromise model effectiveness.</li>
    <li>To combat this issue, we will utilize the <strong>Synthetic Minority Over-sampling Technique (SMOTE)</strong>, which offers several advantages:</li>
    <ul>
      <li><strong>Equal Representation:</strong> SMOTE generates synthetic samples for the minority class, ensuring both classes are equally represented in the dataset.</li>
      <li><strong>Improved Learning:</strong> By balancing the classes, SMOTE allows models to learn more effectively from both classes, enhancing their ability to make accurate predictions.</li>
      <li><strong>Reduction of Bias:</strong> Addressing class imbalance helps mitigate the bias that can occur in models trained on imbalanced datasets, leading to fairer predictions.</li>
      <li><strong>Enhanced Model Performance:</strong> Balanced classes often result in improved metrics, such as accuracy, precision, and recall, providing a more reliable assessment of model performance.</li>
    </ul>
    <li>Overall, employing SMOTE is crucial for improving the robustness and reliability of our machine learning models, ultimately leading to better predictions of water potability.</li>
  </ul>

</div>


```python
# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

#  Print the class distribution after SMOTE
print("Class distribution after SMOTE:")
print(y_resampled.value_counts())
```

Applying __`SMOTE`__ (Synthetic Minority Over-sampling Technique) is useful because it effectively balances class distributions in imbalanced datasets, ensuring that both classes (e.g., Potable and Not Potable) are equally represented. This balance enhances model training by providing the algorithm with enough data to learn from both classes, reducing bias toward the majority class and improving overall prediction accuracy. As a result, models trained on SMOTE-processed data are better equipped to generalize to unseen instances, making them more reliable in real-world applications.

```python
# Check against WHO limits
def check_water_quality(input_data):
    # Unpack the input data (assuming the order of features)
    pH, hardness, solids, chloramines, sulfate, conductivity, organic_carbon, trihalomethanes, turbidity = input_data
    
    # Check WHO limits
    if (6.5 <= pH <= 8.5 and
        hardness <= 200 and
        solids <= 1000 and
        chloramines <= 4 and
        sulfate <= 1000 and
        conductivity <= 400 and
        organic_carbon <= 10 and
        trihalomethanes <= 80 and
        turbidity <= 5):
        return "The water is Potable."
    else:
        return "The water is Not Potable."

# Example: Provide the actual input_data scaled back to its real values
input_data = [7.5, 150.0, 300.0, 3.0, 250.0, 300.0, 3.5, 75.0, 2.0]  # Example input values

# Get the result
result = check_water_quality(input_data)
print(result)
```
